# 特徴量分析
生成した特徴量の相関・重要度・分布を確認する

```python
import sys
sys.path.append('..')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import yaml

with open('../config.yaml') as f:
    config = yaml.safe_load(f)

from src.features.feature_pipeline import load_features, FEATURE_COLUMNS
df = load_features(config)
print(df.shape)
```

```python
# 特徴量と勝率の相関
df['is_win'] = (df['finish'] == 1).astype(int)
corr = df[FEATURE_COLUMNS + ['is_win']].corr()['is_win'].drop('is_win').sort_values()

fig, ax = plt.subplots(figsize=(8, 10))
corr.plot(kind='barh', ax=ax)
ax.set_title('各特徴量と「1着」の相関係数')
ax.axvline(0, color='black', linewidth=0.8)
plt.tight_layout()
plt.show()
```

```python
# 特徴量の相関行列（ヒートマップ）
feat_corr = df[FEATURE_COLUMNS].corr()
plt.figure(figsize=(14, 12))
sns.heatmap(feat_corr, cmap='RdBu_r', center=0, fmt='.1f', annot=False)
plt.title('特徴量間の相関行列')
plt.tight_layout()
plt.show()
```
