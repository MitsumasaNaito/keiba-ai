# データ探索
スクレイピングで取得したデータの基本統計量と分布を確認する

```python
import sys
sys.path.append('..')

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import yaml

with open('../config.yaml') as f:
    config = yaml.safe_load(f)

plt.rcParams['font.family'] = 'Hiragino Sans'  # macOS日本語フォント
```

```python
from src.features.raw_loader import RawLoader

loader = RawLoader(config)
races = loader.load_races()
print(f'Total entries: {len(races)}')
print(f'Total races: {races["race_id"].nunique()}')
races.head()
```

```python
# 着順の分布
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

races['finish'].value_counts().sort_index().plot(kind='bar', ax=axes[0])
axes[0].set_title('着順の分布')

races['course_type'].value_counts().plot(kind='pie', ax=axes[1], autopct='%1.1f%%')
axes[1].set_title('芝/ダート比率')

races['distance'].plot(kind='hist', bins=20, ax=axes[2])
axes[2].set_title('距離の分布')

plt.tight_layout()
plt.show()
```

```python
# 欠損値の確認
missing = races.isnull().sum() / len(races) * 100
missing[missing > 0].sort_values(ascending=False).plot(kind='bar', figsize=(12, 4))
plt.title('欠損率 (%)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
